# Cohort group metadata quick demo

This notebook shows how to load cohort-group metadata tables and filter them with `cohort_group_query`.

```python
from malariagen_data import ag3 as _ag3
from malariagen_data.anoph.cohort_group_metadata import AnophelesCohortGroupMetadata
```

```python
cohort_meta = AnophelesCohortGroupMetadata(
    url="simplecache::gs://vo_agam_release_master_us_central1",
    public_url="gs://vo_agam_release_master_us_central1/",
    config_path=_ag3.CONFIG_PATH,
    major_version_number=_ag3.MAJOR_VERSION_NUMBER,
    major_version_path=_ag3.MAJOR_VERSION_PATH,
    pre=True,
    simplecache={"cache_storage": "../gcs_cache"},
)
cohort_meta
```

```python
df_admin1_month = cohort_meta.cohort_group_metadata("admin1_month")
df_admin1_month.head()
```

```python
df_bf = cohort_meta.cohort_group_metadata(
    "admin1_month",
    cohort_group_query="country == 'Burkina Faso' and year >= 2010",
)
df_bf.loc[:, ["cohort_id", "cohort_size", "country", "year", "month", "admin1_name"]].head()
```

```python
df_bf.groupby(["year", "month"], as_index=False)["cohort_size"].sum().head()
```
