### Parser Architecture

The parser uses keyword matching and pattern extraction to:
1. **Identify the task** (sample_metadata, PCA, SNP frequencies, FST, map visualization)
2. **Extract entities** (countries, years, species, genes)
3. **Build structured parameters** for API calls
4. **Generate API function calls** with appropriate parameters

This approach offers:
- Simple and rule-based (no ML models needed initially)
- Fast and deterministic
- Easy to extend with new keywords
- Interpretable (can explain what was parsed)
- Limitations: Keyword/pattern detection only (future: ontology + entity recognition)
