# Visualizing mosquito sample geographic distribution

This notebook is a **beginner-friendly** introduction to exploring where
mosquito samples were collected using the `malariagen_data` package.

**What you will learn:**

1. How to load sample metadata with `malariagen_data`.
2. How to summarize sample counts by country.
3. How to create a bar chart showing the number of samples per country.
4. How to visualize sample locations on an interactive map.

> **Prerequisites:** A working Python environment with
> `malariagen_data` installed. See the
> [README](https://github.com/malariagen/malariagen-data-python#readme)
> for installation instructions.

## 1. Setup

First we import `malariagen_data` and create a connection to the
**Ag3** dataset — the *Anopheles gambiae* genomic observatory data
hosted on Google Cloud Storage.

```python
import malariagen_data

ag3 = malariagen_data.Ag3(
    "simplecache::gs://vo_agam_release_master_us_central1",
    simplecache=dict(cache_storage="../gcs_cache"),
)
ag3
```

## 2. Load sample metadata

The `sample_metadata()` method returns a pandas DataFrame with one row
per mosquito sample. It includes columns such as `country`, `location`,
`latitude`, `longitude`, `year`, `taxon`, and more.

```python
df_samples = ag3.sample_metadata()
print(f"Total samples: {len(df_samples):,}")
print(f"Columns: {df_samples.columns.tolist()}")
df_samples.head()
```

## 3. Summarize samples by country

Let's count how many samples were collected in each country and display
the results as a sorted table.

```python
country_counts = (
    df_samples["country"]
    .value_counts()
    .reset_index()
)
country_counts.columns = ["country", "sample_count"]
country_counts
```

## 4. Bar chart — samples by country

The `plot_samples_bar()` method creates an interactive Plotly bar chart.
Pass `x="country"` to see sample counts per country.

```python
ag3.plot_samples_bar(x="country")
```

We can also color the bars by `taxon` to see which mosquito species
are present in each country:

```python
ag3.plot_samples_bar(x="country", color="taxon")
```

## 5. Interactive map

The `plot_samples_interactive_map()` method displays collection
locations on an interactive [Leaflet](https://leafletjs.com/) map.
Each marker represents a sampling location, with the size proportional
to the number of samples collected there.

```python
ag3.plot_samples_interactive_map()
```

## 6. Exploring by year

We can also look at how sampling effort has changed over time by
plotting samples by year:

```python
ag3.plot_samples_bar(x="year", color="country")
```
