## 🌍 What is MalariaGEN?

**MalariaGEN** (Malaria Genomic Epidemiology Network) is a global collaboration of researchers studying malaria parasites and the mosquitoes that transmit them.

### Why This Matters:
- 🦟 **Mosquitoes** transmit malaria to humans
- 🩸 **Parasites** cause the disease in humans
- 🧬 **Genomics** helps us understand how to fight malaria
- 🌍 **Global data** helps track drug resistance and disease spread

### Available Data Types:
- **Anopheles mosquitoes** (vectors): Ag3, Af1, Amin1, Adir1, Adar1
- **Plasmodium parasites** (pathogens): Pf7, Pf8, Pv4

Let's start exploring!

## 📦 Setting Up Your Environment

```python
# First, let's install and import the malariagen_data package
# If you're running this in Google Colab, this should work automatically
# If you're running locally, you might need: !pip install malariagen_data

import malariagen_data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set up nice plotting style
plt.style.use('default')
sns.set_palette("husl")

print("✅ All packages imported successfully!")
print(f"📊 malariagen_data version: {malariagen_data.__version__}")
```

## 🔍 Exploring Available Data Resources

```python
# Let's see what data resources are available
print("🌍 Available MalariaGEN Data Resources:\n")

# Mosquito data (vectors)
print("🦟 MOSQUITO DATA (Vectors):")
print("  • Ag3 - Anopheles gambiae complex (major African malaria vectors)")
print("  • Af1 - Anopheles funestus (important African vector)")
print("  • Amin1 - Anopheles minimus (Southeast Asian vector)")
print("  • Adir1 - Anopheles dirus complex (Southeast Asian vector)")
print("  • Adar1 - Anopheles dirus (additional Southeast Asian data)")

print("\n🩸 PARASITE DATA (Pathogens):")
print("  • Pf7 - Plasmodium falciparum (most deadly malaria parasite)")
print("  • Pf8 - Plasmodium falciparum (latest release)")
print("  • Pv4 - Plasmodium vivax (causes relapsing malaria)")

print("\n🚀 Let's start with the most popular dataset: Ag3 (African mosquitoes)")
```

## 🦟 Exploring Mosquito Data (Ag3)

```python
# Initialize the Ag3 dataset (Anopheles gambiae)
# This will connect to the MalariaGEN data repository
ag3 = malariagen_data.Ag3()

print("✅ Ag3 dataset initialized successfully!")
print("📊 This dataset contains genomic data from African malaria mosquitoes")
```

```python
# Let's explore what's available in this dataset
print("🔍 Available releases:")
releases = ag3.releases()
print(releases)

print(f"\n📈 Total releases available: {len(releases)}")
```

```python
# Let's look at sample sets (groups of mosquito samples)
print("🧬 Available sample sets:")
sample_sets = ag3.sample_sets()
print(f"Total sample sets: {len(sample_sets)}")

# Show first few sample sets
print("\nFirst 5 sample sets:")
print(sample_sets.head())
```

## 🌍 Exploring Sample Metadata

```python
# Let's get sample metadata (information about where/when mosquitoes were collected)
print("📍 Loading sample metadata...")
sample_metadata = ag3.sample_metadata()

print(f"📊 Total samples in dataset: {len(sample_metadata):,}")
print(f"📋 Columns available: {list(sample_metadata.columns)}")
```

```python
# Let's explore the geographical distribution
print("🌍 Geographical Distribution of Samples:")

# Count samples by country
country_counts = sample_metadata['country'].value_counts().head(10)
print("\nTop 10 countries with most samples:")
for country, count in country_counts.items():
    print(f"  {country}: {count:,} samples")
```

```python
# Create a simple visualization of sample distribution
plt.figure(figsize=(12, 6))

# Plot top 15 countries
top_countries = sample_metadata['country'].value_counts().head(15)

plt.barh(range(len(top_countries)), top_countries.values)
plt.yticks(range(len(top_countries)), top_countries.index)
plt.xlabel('Number of Samples')
plt.title('🌍 MalariaGEN Ag3: Top 15 Countries by Sample Count')
plt.gca().invert_yaxis()  # Put highest count at top

# Add value labels on bars
for i, (country, count) in enumerate(top_countries.items()):
    plt.text(count + max(top_countries) * 0.01, i, f'{count:,}', 
             va='center', fontweight='bold')

plt.tight_layout()
plt.show()

print("📊 This chart shows which countries have contributed the most mosquito samples to MalariaGEN!")
```

## 🧬 Understanding Mosquito Species

```python
# Let's look at the different mosquito species in the dataset
print("🦟 Different mosquito species in the dataset:")

if 'taxon' in sample_metadata.columns:
    species_counts = sample_metadata['taxon'].value_counts()
    print("\nSpecies distribution:")
    for species, count in species_counts.items():
        print(f"  {species}: {count:,} samples")
else:
    print("Taxon information not available in this dataset")
    
    # Let's check what columns we do have
    print("\nAvailable columns for species information:")
    species_cols = [col for col in sample_metadata.columns if 'species' in col.lower() or 'taxon' in col.lower()]
    print(f"Species-related columns: {species_cols}")
```

## 📅 Exploring Collection Timeline

```python
# Let's see when these samples were collected
print("📅 Exploring collection timeline...")

# Check if we have year information
year_cols = [col for col in sample_metadata.columns if 'year' in col.lower()]
print(f"Year-related columns: {year_cols}")

if year_cols:
    year_col = year_cols[0]
    year_counts = sample_metadata[year_col].value_counts().sort_index()
    
    print(f"\n📊 Sample collection by {year_col}:")
    for year, count in year_counts.items():
        print(f"  {year}: {count:,} samples")
        
    # Plot timeline
    plt.figure(figsize=(12, 6))
    plt.plot(year_counts.index, year_counts.values, marker='o', linewidth=2, markersize=8)
    plt.xlabel('Year')
    plt.ylabel('Number of Samples')
    plt.title('📅 MalariaGEN Ag3: Sample Collection Timeline')
    plt.grid(True, alpha=0.3)
    plt.show()
else:
    print("No year information available in this dataset")
```

## 🩸 Exploring Parasite Data (Pf7)

```python
# Now let's explore parasite data (Plasmodium falciparum)
print("🩸 Initializing Pf7 dataset (Plasmodium falciparum parasites)...")

pf7 = malariagen_data.Pf7()
print("✅ Pf7 dataset initialized successfully!")
print("🦠 This dataset contains genomic data from malaria parasites")
```

```python
# Get parasite sample metadata
print("🧬 Loading parasite sample metadata...")
parasite_metadata = pf7.sample_metadata()

print(f"📊 Total parasite samples: {len(parasite_metadata):,}")
print(f"📋 Available columns: {list(parasite_metadata.columns)}")
```

```python
# Explore geographical distribution of parasites
print("🌍 Parasite geographical distribution:")

if 'country' in parasite_metadata.columns:
    parasite_countries = parasite_metadata['country'].value_counts().head(10)
    print("\nTop 10 countries for parasite samples:")
    for country, count in parasite_countries.items():
        print(f"  {country}: {count:,} samples")
        
    # Compare mosquito vs parasite distribution
    plt.figure(figsize=(14, 8))
    
    # Get top countries for both
    mosquito_top = sample_metadata['country'].value_counts().head(8)
    parasite_top = parasite_metadata['country'].value_counts().head(8)
    
    # Create side-by-side comparison
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    
    # Mosquito samples
    ax1.barh(range(len(mosquito_top)), mosquito_top.values, color='skyblue')
    ax1.set_yticks(range(len(mosquito_top)))
    ax1.set_yticklabels(mosquito_top.index)
    ax1.set_xlabel('Number of Samples')
    ax1.set_title('🦟 Mosquito Samples (Ag3)')
    ax1.invert_yaxis()
    
    # Parasite samples
    ax2.barh(range(len(parasite_top)), parasite_top.values, color='lightcoral')
    ax2.set_yticks(range(len(parasite_top)))
    ax2.set_yticklabels(parasite_top.index)
    ax2.set_xlabel('Number of Samples')
    ax2.set_title('🩸 Parasite Samples (Pf7)')
    ax2.invert_yaxis()
    
    plt.tight_layout()
    plt.show()
    
    print("📊 This comparison shows how mosquito and parasite data collection differs by region!")
else:
    print("Country information not available in parasite dataset")
```
