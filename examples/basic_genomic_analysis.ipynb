# Exploring MalariaGEN Mosquito Genomic Data

This notebook demonstrates how to access mosquito genomic
metadata using the malariagen_data API and perform simple
exploratory analysis.

The analysis includes:

- Loading sample metadata
- Exploring country distribution
- Visualizing mosquito species counts

```python
import malariagen_data
import pandas as pd
import plotly.express as px
```

```python
# Initialize the MalariaGEN API
ag3 = malariagen_data.Ag3()

print("API initialized successfully")
```

```python
# Load mosquito sample metadata
samples = ag3.sample_metadata()

print("Total samples:", len(samples))

samples.head()
```

The dataset contains metadata for mosquito samples collected
from multiple countries.

Important fields include:

- country
- location
- year
- taxon (mosquito species)

```python
country_counts = samples["country"].value_counts().reset_index()

country_counts.columns = ["country", "sample_count"]

country_counts.head()
```

```python
fig = px.bar(
    country_counts.head(15),
    x="country",
    y="sample_count",
    title="Top Countries by Mosquito Sample Count"
)

fig.show()
```

```python
uganda_samples = samples[samples["country"] == "Uganda"]

print("Samples from Uganda:", len(uganda_samples))

uganda_samples.head()
```

```python
species_counts = samples["taxon"].value_counts().reset_index()

species_counts.columns = ["species", "count"]

species_counts
```

```python
fig = px.pie(
    species_counts.head(10),
    values="count",
    names="species",
    title="Mosquito Species Distribution"
)

fig.show()
```
